Visualization

```python
# PAIRPLOT
sns.pairplot(df[['Age','Fare','Survived']],hue='Survived')# Making the pairplot
# Passengers with higher fare and lower age tend to have a higher chance of survival.
# Survived = 1 (green or blue points) are mostly clustered around low age and high fare regions.
```

```python
# HEATMAP
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm')# Making the heatmap
plt.title("Correlation Heatmap")
plt.show()
#Fare aur Survived ke beech moderate positive correlation hai,
#matlab mehenga ticket lene wale passengers zyada survive hue.Pclass ka Survived ke saath negative correlation hai,
#yani 3rd class mein survival rate kam tha.
```

Plots for data visualization

```python
# SCATTERPLOT
sns.scatterplot(data=df, x='Age', y='Fare', hue='Survived')# Making the scatterplot
plt.title("Age vs Fare (colored by Survival)")
plt.show()
# The chart shows that younger passengers (ages 0â€“20) who paid higher fares had better chances of survival (more orange dots).
# Most of the passengers who did not survive (blue dots) are in the low fare and mid-age range.
```

```python
df[['Age', 'Fare']].isnull().sum()
```

```python
# HISTPLOT
sns.histplot(data=df, x='Age', bins=30, kde=True)# Making the histplot
plt.title("Age Distribution")
plt.show()
# Most passengers were between the ages of 20 and 40.
# Very few people belonged to the 60+ age group.
```

```python
# BOXPLOT BY AGE AND PASSENGER CLASS
sns.boxplot(data=df, x='Pclass', y='Age')# Making the boxplot.
plt.title("Age by Passenger Class")
plt.show()
# The average age of survived passengers is slightly lower.
# Young passengers, especially children, had a higher chance of survival.
```

```python
# GROUP BY SURVIVAL RATE
df.groupby('Pclass')['Survived'].mean().plot(kind='bar')# using the groupby in Pclass and Survived columns.
plt.title('Survival Rate by Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
# 1st class passengers had the highest survival rate.
# 3rd class passengers had the lowest survival rate, highlighting the socio-economic gap.
```

```python
# BOXPLOT BY FARE AND PCLASS
sns.boxplot(x='Pclass', y='Fare', data=df)
plt.title("Fare Distribution by Passenger Class")
plt.show()
# 1st class passengers paid the highest fares, with several outliers observed.
# Fares in 3rd class were much lower and had a narrower range.
```

```python
# GROUP BY SURVIVAL RATE AND GENDER 
df.groupby('Sex')['Survived'].mean().plot(kind='bar')
plt.title('Survival Rate by Gender')
plt.ylabel('Survival Rate')
plt.show()
# The survival rate of women was significantly higher compared to men.
#The impact of the Titanic's rescue policy of 'women and children first' is clearly visible.
```
